```python
scaler = StandardScaler()

X1_train_scaled = scaler.fit_transform(X1_train)
X1_test_scaled = scaler.fit_transform(X1_test)
```

# Adjusted R-Squared calculation

Instead of calculating R-Squared, we can use adjusted R-Squared, this is more suitable for a large dataset.

#### Formula for adjusted R-Squared

$$R_(adj)^ 2 = 1 - {\frac{(1-R^2)(n-1)}{n-k-1}}$$

- n = No. of data
- k = No. of variables

```python
n = 1536
```

```python
k = 112
```

# Linear Regression model

```python
lr = LinearRegression()
lr.fit(X1_train_scaled, y1_train)
```

```python
# Adjusted R-Squared for Linear Regression model
r2_lr = lr.score(X1_train_scaled , y1_train)
adj_r2_lr = 1 - ((1 - r2_lr) * (n - 1) / (n - k -1))

# RMSE for Linear Regression model
rmse_lr = -cross_val_score(lr, X1_train_scaled , y1_train , cv=5 , scoring ='neg_root_mean_squared_error').mean()
```

```python
print(f'Adjusted R2 for Linear Regression: {adj_r2_lr}')
print(f'RMSE for Linear Regression: {rmse_lr}')
```

# Ridge CV model

```python
r_alphas = np.linspace(10, 30, 200)

rid_cv = RidgeCV(alphas = r_alphas , cv=5)
rid_cv = rid_cv.fit(X1_train_scaled , y1_train)

# Adjusted R-Squared for Ridge CV model
r2_rid_cv = rid_cv.score(X1_train_scaled , y1_train)
adj_r2_rid_cv = 1 - ((1 - r2_rid_cv) * (n - 1) / (n - k -1))

# RMSE for Ridge CV model
rmse_rid_cv = -cross_val_score(rid_cv, X1_train_scaled , y1_train , cv=5 , scoring ='neg_root_mean_squared_error').mean()
```

```python
print(f'Adjusted R2 for Ridge CV: {adj_r2_rid_cv}')
print(f'RMSE for Ridge CV: {rmse_rid_cv}')
```

# Lasso CV model

```python
l_alphas = np.linspace(150 , 220 , 150)

las_cv = LassoCV(alphas=l_alphas , cv=5 , max_iter = 1000)
las_cv = las_cv.fit(X1_train_scaled , y1_train)

# Adjusted R-Squared for Lasso CV model
r2_las_cv = las_cv.score(X1_train_scaled , y1_train)
adj_r2_las_cv = 1 - ((1 - r2_las_cv) * (n - 1) / (n - k -1))

# RMSE for Lasso CV model
rmse_las_cv = -cross_val_score(las_cv, X1_train_scaled , y1_train , cv=5 , scoring ='neg_root_mean_squared_error').mean()
```

```python
print(f'Adjusted R2 for Lasso CV: {adj_r2_las_cv}')
print(f'RMSE for Lasso CV: {rmse_las_cv}')
```
