# Data Cleaning

Check for Null Values.

```python
# Check for Null Values
test.isnull().sum()
```

We will apply the similar techniques as our train datasets, we will drop the following variables: `Alley`, `Fireplace Qu`, `Pool QC`, `Fence` & `Misc Feature`

```python
test.drop(columns=['Alley' , 'Fireplace Qu' , 'Pool QC' , 'Fence' , 'Misc Feature'],axis=1,inplace=True)
```

```python
# Check for nulls after dropping 5 columns
test.isnull().sum()
```

Filling up those missing values, with less than 1000 null:

- Replace those numerical values by its mode
- Replace those non-numerical values by the maximum occurencce

```python
# Check those numerical null values
(test.isnull().sum() > 0) & (test.dtypes != object)
```

```python
# Replace those numerical values by its mode
test['Lot Frontage'] = test['Lot Frontage'].fillna(test['Lot Frontage'].mode()[0])
test['Mas Vnr Area'] = test['Mas Vnr Area'].fillna(test['Mas Vnr Area'].mode()[0])
test['Garage Yr Blt'] = test['Garage Yr Blt'].fillna(test['Garage Yr Blt'].mode()[0])
```

```python
# Check those non-numerical null values
(test.isnull().sum() > 0) & (test.dtypes == object)
```

```python
# Check the number of elements inside 'Mas Vnr Type'
test.loc[:,'Mas Vnr Type'].value_counts(dropna=False)
```

```python
# Replace the missing values with None (as this is the maximum)
test.loc[:,'Mas Vnr Type'] = test.loc[:,'Mas Vnr Type'].fillna('None')
test.loc[:,'Mas Vnr Type'].value_counts(dropna=False)
```

```python
# Check the number of elements inside 'Bsmt Qual'
test.loc[:,'Bsmt Qual'].value_counts(dropna=False)
```
