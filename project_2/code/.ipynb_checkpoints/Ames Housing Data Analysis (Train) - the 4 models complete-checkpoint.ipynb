# Scaling data

```python
scaler = StandardScaler()

X1_train_scaled = scaler.fit_transform(X1_train)
X1_test_scaled = scaler.fit_transform(X1_test)
```

# Adjusted R-Squared calculation

Instead of calculating R-Squared, we can use adjusted R-Squared, this is more suitable for a large dataset.

#### Formula for adjusted R-Squared

$$R_(adj)^ 2 = 1 - {\frac{(1-R^2)(n-1)}{n-k-1}}$$

- n = No. of data
- k = No. of variables

# Linear Regression model

```python
lr = LinearRegression()
```

```python
lr.fit(X1_train_scaled, y1_train)
```

```python
# Train score (R2) for Linear Regression model on seen data
lr.score(X1_train_scaled , y1_train)
```

```python
# r2 for Linear Regression
lr_r2 = cross_val_score(lr, X1_train_scaled, y1_train, cv=5, scoring='r2').mean()
```

```python
lr_r2
```

```python
# Train score (R2) for Linear Regression model
cross_val_score(lr, X1_train_scaled, y1_train, cv=5).mean()
```

```python
# Train score (R2) for Linear Regression model on unseen data
cross_val_score(lr , X1_train_scaled , y1_train , cv=5)
```

```python
# Estimated RMSE for Linear Regression model
-cross_val_score(lr, X1_train_scaled , y1_train , cv=5 , scoring ='neg_root_mean_squared_error').mean()
```
