#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
def get_standard_deviation(data_list , mean):
    return (sum([(data_point - mean) ** 2 for data_point in data_list]) / len(data_list)) ** 0.5
    
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
#Build dictionary of manually calculated standard deviations
stddev_dict = {column_name: get_standard_deviation(final.loc[:,column_name].tolist(), final.loc[:,column_name].mean()) for column_name in final.columns if column_name != 'state'}
stddev_dict_df = pd.DataFrame(stddev_dict , index=['manual_stddev']).transpose()
````

````python
#Build dictionary of numpy calculated standard deviations
numpy_stddev_dict = {column_name: np.std(final.loc[:,column_name].tolist()) for column_name in final.columns if column_name != 'state'}
numpy_stddev_dict = pd.DataFrame(numpy_stddev_dict , index=['numpy_stddev']).transpose()
````

````python
# merge the above 2 standard deviation dataframes with the pandas describe() dataframe for final comparison
interim = pd.merge(stats_describe.loc[:,['std']].rename(columns={'std':'pandas_stddev'}), stddev_dict_df, left_index=True, right_index=True)
pd.merge(interim, numpy_stddev_dict, left_index=True, right_index=True)
````

Do your manually calculated standard deviations match up with the output from pandas `describe`? What about numpy's `std` method?

By looking at the standard deviation we have calculated based on Pandas, Manual Formula & Numpy, we see that the Manual Formula is same as Numpy result.

The Pandas standard deviation is slightly larger, because it used the sample standard deviation formula.

#### Pandas use the sample standard deviation formula

$$\sigma = \sqrt{\frac{1}{n - 1}\sum_{i=1}^n(x_i - \mu)^2}$$

Compared to the earlier formula given, which is for population standard deviation, the difference lies in the denominator. Population standard deviation formula assumes full population size to be known at computation time, hence N is used. However, for sample standard deviation, N - 1 need to used in order for it to be an unbiased estimator of the population. Once the necessary adjustent is made to the denominator of the manual and numpy formulas, the standard deviations values would be similar to the pandas-derived values.

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, fig_size, list_of_xlabels, list_of_ylabels, ylimits):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=fig_size) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column], bins=11) # feel free to add more settings
        ax[i].set_title(list_of_titles[i], fontsize=18)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=8)
        ax[i].set_ylabel(list_of_ylabels[i], fontsize=8)
        ax[i].set_ylim(ylimits[i]) # otherwise it would have the .5 segments
````
