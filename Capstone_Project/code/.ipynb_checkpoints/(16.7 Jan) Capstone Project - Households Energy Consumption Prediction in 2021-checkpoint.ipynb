### 2.1. Definition of sunshine hours
[top](#Contents)

Given the theoretical maximum of daytime duration for a given location, there is also a practical consideration at which point the amount of daylight is sufficient to be treated as a "sunshine hour". "Bright" sunshine hours represent the total hours when the sunlight is stronger than a specified threshold, as opposed to just "visible" hours. "Visible" sunshine, for example, occurs around sunrise and sunset, but is not strong enough to excite the sensor.

<img src = "../image/Sunshine.PNG">

According to Wikipedia, Singapore has a sunshine duration of about 129.2 - 192.7 hours per month. It is a general indicator of cloudines of a location. It is equivalent to 4.3 - 6.4 hours per day, which match the result in the `total_household_electricity_consumption_2005_2020.csv` data.

## 3. Clean the data
[top](#Contents)

### 3.1. Clean the data
[top](#Contents)

We will first check the duration of the data given.

#### 1. `sunshine` data.

```python
sunshine['month'].unique()
```

For sunshine data, we are given from 1982-01 to 2020-10, all in order. Since we only have household data from 2005 onwards, and we do not have full year data for 2020, so we will extract only data from 2005-01 to 2019-12.

```python
# Extract result from 2005-01 to 2019-12
sunshine_extract = sunshine[(sunshine['month'] > '2004-12') & (sunshine['month'] < '2020-01')]
sunshine_extract.head()
```

```python
# Set the month as index
sunshine_extract.set_index('month', inplace=True)
sunshine_extract.head()
```

#### 2. `household` data.

```python
household['month'].unique()
```

For household data, we are given data from 2005-01 to 2020-06, not in order. We will sort the data first, since we do not have full year data for 2020, so we will extract only data from 2005-01 to 2019-12.

```python
# Sort the result
household  = household.sort_values('month')
household = household.reset_index(drop=True)
```

```python
# Extract restult till 2019-12
household_extract = household[(household['month'] < '2020-01')]
household_extract.head()
```

```python
# Set the month as index
household_extract.set_index('month', inplace=True)
household_extract.head()
```

### 3.2. Combine the data
[top](#Contents)

```python
energy_final = pd.concat([sunshine_extract , household_extract] , axis=1)
```

```python
energy_final.head()
```

```python
energy_final.shape
```

## 6. Modeling 
[top](#Contents)

Modeling with time series data presents unique challenges because our observations are not independent of one another. 

So, in splitting the data into `train` and `test` sets, **we cannot split randomly**. Use the first *75%* of observations as training data and the following 25% for test data.

```python
print(f'Number of observations: {len(energy_final)}')
print(f'Number of training observations: {round(len(energy_final) * 0.75)}')
print(f'Number of testing observations: {len(energy_final) - round(len(energy_final) * 0.75)}')
```

```python
def train_test_plot(train, test):
    plt.figure(figsize=(30, 12))
    plt.plot(train, c='blue')
    plt.plot(test, c='orange');
    plt.xlabel('Time',fontsize=18)
    plt.xticks(rotation=90, fontsize=6)
    plt.ylabel('Electricity Consumption in KWh',fontsize=18)
```

### 6.1. Train/Test split for 1 room & 2 room data
[top](#Contents)

```python
train_1room_2room = energy_final['1-room_2-room'].iloc[:134,]
test_1room_2room = energy_final['1-room_2-room'].iloc[134:,]
```

```python
train_test_plot(train_1room_2room , test_1room_2room)
```

### 6.2. Train/Test split for 3 room data
