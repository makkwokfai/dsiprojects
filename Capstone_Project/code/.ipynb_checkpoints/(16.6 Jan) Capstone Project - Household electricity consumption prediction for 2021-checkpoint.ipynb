## 2. Check the data given
[top](#Contents)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
household = pd.read_csv('../datasets/total_household_electricity_consumption_2005_2020.csv')
household.head()
```

```python
household.tail()
```

```python
household.describe()
```

## 3. Clean the data
[top](#Contents)

We will first check the duration of the data given.

```python
household['month'].unique()
```

#### Interpretation:

We are given monthly result from 2005-01 to 2020-07, not in order. We will sort the data so that we may predict the future result using the last 1 year data.

```python
# Sort the result
household  = household.sort_values('month')
household = household.reset_index(drop=True)
```

```python
# Confirm the household month result is sorted appropriately.
household['month'].unique()
```

```python
# Extract the result from July 2019
household_from_Jul19 = household[household['month'] > '2019-06']
```

```python
# Set month as the index
household_from_Jul19.set_index('month', inplace=True)
household_from_Jul19.head()
```

```python
household_from_Jul19.tail()
```

## 6. Increased household electricity
[top](#Contents)

<img src = "../image/Increased household electricity.PNG">

As the circuit breaker begins in April, the parliament mentioned that there will be an increased in household electricity from April to July.

That is a reasonable statement. However, our data only shows a surge in electricity consumption starting from June. It is possible that there is a lag in the data we have collected. We will next look at the percentage increased in the electricity used.

```python
household_from_Jul19.pct_change().tail()
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['1-room_2-room'].pct_change().plot()
plt.title('1-room & 2-room electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['3-room'].pct_change().plot()
plt.title('3-room electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['4-room'].pct_change().plot()
plt.title('4-room electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['5-room_and_executive'].pct_change().plot()
plt.title('5-room & executive electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['private_apts_and_condo'].pct_change().plot()
plt.title('Private apartments & condo electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['landed_properties'].pct_change().plot()
plt.title('Landed properties electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
plt.figure(figsize=(12, 8))
household_from_Jul19['others'].pct_change().plot()
plt.title('Others electricity consumption percentage changes')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

```python
household_from_Jul19.pct_change().plot(figsize=(12,8))
plt.title('Electricity consumption percentage changes for all types of household')
plt.xlabel('Time')
plt.ylabel('Percentage change in eletricity consumption');
```

#### Percentage changes in electricity consumption after circuit breaker:

- 1-room & 2-room: 26.2%
- 3-room: 28.8%
- 4-room: 30.6%
- 5-room & executive: 33.9%
- Private apartments & condo: 37.7%
- Landed properties: 17.7%
- Others: 10%
